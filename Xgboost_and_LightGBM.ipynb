```python
import numpy as np
np.unique(y_train)
```

```python
from xgboost import XGBClassifier
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
params = {
        "n_estimators": [50,100,150,200], #no. of trees
        "max_depth" : [3, 4, 5, 7], #depth of each tree
        "learning_rate": [0.1, 0.2, 0.3], #eta or shrinkage
        'subsample': [0.6, 0.8, 1.0], #row sample
        'colsample_bytree': [0.6, 0.8, 1.0], #column sampling
        }
```

```python
xgb= XGBClassifier(objective='multi:softmax', num_class=20)

random_search= RandomizedSearchCV(xgb,param_distributions= params, scoring='accuracy',
                   n_jobs=-1, cv=3, verbose=1, n_iter=10)
```

```python
random_search.fit(X_train, y_train)
```

```python
xgb= random_search.best_estimator_
```

```python
xgb.fit(X_train, y_train)
```

```python
xgb.score(X_test, y_test)
```

```python
import lightgbm as lgb
Params = {
    'learning_rate': [0.1, 0.3, 0.5],
    'boosting_type' : ['gbdt'],
    'objective' : ['multiclass'],
    'max_depth' : [5,6,7,8],
    'colsample_bytree' : [0.5,0.7],
    'subsample' : [0.5,0.7]
    }
```

```python
clf = lgb.LGBMClassifier(num_classes=20)
```

```python
random_search_lgbm= RandomizedSearchCV(clf,Params, scoring='accuracy',n_jobs=-1, cv=3, verbose=1, n_iter=10)
```

```python
random_search_lgbm.fit(X_train,y_train)
```

```python
lightgbm_clf= random_search_lgbm.best_estimator_
```

```python
lightgbm_clf.fit(X_train, y_train)
```

```python
lightgbm_clf.score(X_test, y_test)
```
